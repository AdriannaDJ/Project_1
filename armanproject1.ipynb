```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

#File to Load 
Netflix_data_to_load = pd.read_csv ("netflix_titles 2.csv") 

# Display sample data
Netflix_data_to_load.head
```

```python
#count NA values
null_counts = Netflix_data_to_load.isnull().sum()
print(null_counts)
```

```python
# Drop NA CounNetflix_data_to_loadtry 
# Drop rows with any missing values

Clean_Data = Netflix_data_to_load.dropna(subset=['country','rating'])

#Clean_Data = Netflix_data_to_load.isnull().sum() ##Used for testing

Clean_Data
```

```python
#New DF with paired down columns related to questions

Cleaned_Data = Clean_Data.loc[:, ["type", "title", "country", "date_added", "release_year", "rating", "duration", "listed_in"]]
Cleaned_Data
```

```python
# Splitting the date_added column into month, day, and year
Cleaned_Data[['month/day_added', 'year_added']] = Cleaned_Data['date_added'].str.split(',', expand=True)

# Displaying the updated DataFrame
Cleaned_Data
```

```python
#breaking down the genres
categories_df=Cleaned_Data['listed_in'].str.get_dummies(', ')
categories=list(categories_df.columns)
display(len(categories))
Cleaned_Data[categories]=categories_df
Cleaned_Data['num_categories']=categories_df.sum(axis=1)
Cleaned_Data
```

```python
#Movies and TV Shows produced in Countries by size. Top 10.
newt = Cleaned_Data.groupby('country').size().sort_values(ascending=False).head(10)
newt = newt.index.tolist()
Top_10_df = Cleaned_Data[Cleaned_Data['country'].isin(newt)]
Top_10_df
```

```python
top_5_categories=list(categories_df.sum().sort_values(ascending=False).head(5).index)
Cleaned_Data[['release_year']+top_5_categories].groupby('release_year').sum().plot()
```

```python
movies_ratings = Top_10_df[Top_10_df['type'] == 'Movie'][['title', 'rating', 'country']]
movie_rating_count = movies_ratings['rating'].count()
movie_rating_perc = round((movies_ratings.groupby(['rating','country']).size().sort_values(ascending=False) / movie_rating_count) * 100, 2)
movie_rating_perc

```

```python
# Calculate the percentage of each rating for each country
movies_ratings = Top_10_df[Top_10_df['type'] == 'Movie'][['title', 'rating', 'country']]
movie_rating_count = movies_ratings['rating'].count()
movie_rating_perc = round((movies_ratings.groupby(['rating', 'country']).size().sort_values(ascending=False) / movie_rating_count) * 100, 2)

# Create a DataFrame to display the information
movie_rating_table = movie_rating_perc.unstack().fillna(0)
print(movie_rating_table)
```

```python
movie_rating_perc_reset = movie_rating_perc.reset_index()

# Pivot the DataFrame to have 'rating' as columns for stacked bar chart
pivot_table = movie_rating_perc_reset.pivot(index='country', columns='rating', values=0).fillna(0)

# Create a stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.xlabel('Country')
plt.ylabel('Percentage')
plt.title('Distribution of Ratings for Movies by Country')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
tv_ratings = Top_10_df[Top_10_df['type'] == 'TV Show'][['title', 'rating']]
tv_ratings = tv_ratings.merge(Top_10_df[['title', 'country']], on='title', how='left')
tv_rating_count = tv_ratings['rating'].count()
tv_rating_perc = round((tv_ratings.groupby(['rating','country']).size().sort_values(ascending=False) / tv_rating_count) * 100, 2)
tv_rating_perc
```

```python
# Calculate the percentage of each rating for each country
tv_ratings = Top_10_df[Top_10_df['type'] == 'TV Show'][['title', 'rating', 'country']]
tv_rating_count = tv_ratings['rating'].count()
tv_rating_perc = round((tv_ratings.groupby(['rating', 'country']).size().sort_values(ascending=False) / tv_rating_count) * 100, 2)

# Create a DataFrame to display the information
tv_rating_table = tv_rating_perc.unstack().fillna(0)
print(tv_rating_table)
```
